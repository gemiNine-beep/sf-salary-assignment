```r
# Set the path to your ZIP file
zip_file_path <- "Employee Profile.zip"

# Create a folder to unzip into
unzip_dir <- "Employee_Profile_extracted"
if (!dir.exists(unzip_dir)) {
  dir.create(unzip_dir)
}

# Unzip the file
unzip(zipfile = zip_file_path, exdir = unzip_dir)
cat("Unzipped successfully into:", unzip_dir, "\n")

# Find the CSV file inside
csv_files <- list.files(path = unzip_dir, pattern = "\\.csv$",
                        full.names = TRUE, recursive = TRUE)
cat("Found CSV file(s):", paste(csv_files, collapse = ", "), "\n\n")

# Read and display the CSV
employee_data <- read.csv(csv_files[1], stringsAsFactors = FALSE)
cat("Employee Details:\n")
print(employee_data)
```
