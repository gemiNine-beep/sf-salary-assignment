## Get Employee by filtering with employee's name

```python
def get_employee_info(employee_name):
  found_employee = df[df.EmployeeName == employee_name]
  if found_employee.empty:
    return "No Employee Found"
  else:
    return found_employee
```

```python
get_employee_info("NATHANIEL FORD")
```

## Process Employee Salary

```python
def get_employee_total_comp(employee_name):
  """ Get total compensation from dataframe"""
  try:
    found_employee = df[df.EmployeeName == employee_name]
    return found_employee[["BasePay", "OvertimePay", "OtherPay", "Benefits", "TotalPay", "TotalPayBenefits"]]
  except TypeError:
    return "No Employee Found"

```

```python
total_salary = get_employee_total_comp("NATHANIEL FORD")
```

```python
total_salary.to_dict()
```

```python
def get_monthly_salary(employee_name):
  try:
    employee_total_comp = get_employee_total_comp(employee_name)
    salary = employee_total_comp["TotalPay"].values[0] / 12
    return dict(name=employee_name,  monthly_pay=salary)
  except IndexError:
    return "No Employee Found"
```

```python
get_monthly_salary("NATHANIEL FORD")
```

## Export Employee Profile to CSV

```python
def export_employee_profile(employee_name):
    try:
        found_employee = get_employee_info(employee_name)

        folder_name = "Employee Profile"
        os.makedirs(folder_name, exist_ok=True)

        csv_path = os.path.join(folder_name, f"{employee_name}.csv")
        found_employee.to_csv(csv_path, index=False)

        shutil.make_archive(folder_name, "zip", folder_name)

        return "Success"

    except AttributeError:
        return "No Employee Found"
```

```python
export_employee_profile("NATHANIEL FORD")
```

```python
import zipfile

## Export Employee Total Compensation to a CSV file inside a ZIP folder

def export_employee_to_zip(employee_name):
  """ Export employee's total compensation to CSV inside a ZIP folder """
  total_comp = get_employee_total_comp(employee_name)

  if isinstance(total_comp, str):
    print("No Employee Found - nothing to export")
    return None

  csv_filename = f"{employee_name.replace(' ', '_')}_details.csv"
  zip_filename = "Employee Profile.zip"

  total_comp.to_csv(csv_filename, index=False)

  with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zf:
    zf.write(csv_filename, arcname=f"Employee Profile/{csv_filename}")

  os.remove(csv_filename)
  print(f"Exported successfully: {zip_filename}")
  return zip_filename


export_employee_to_zip("NATHANIEL FORD")
```
